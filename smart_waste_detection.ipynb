```python
import os

dataset_path = "/content/yolov5/smart-waste-dataset"

for folder in ["images/train", "images/val", "labels/train", "labels/val"]:
    full_path = os.path.join(dataset_path, folder)
    if os.path.exists(full_path):
        print(f"✅ Found: {full_path}")
    else:
        print(f"❌ Missing: {full_path}")
```

```python
yaml_content = """
train: /content/yolov5/smart-waste-dataset/images/train
val: /content/yolov5/smart-waste-dataset/images/val

nc: 2
names: ['organic', 'plastic']
"""

with open("/content/yolov5/smart-waste-dataset/data.yaml", "w") as f:
    f.write(yaml_content)

print("✅ data.yaml created successfully!")
```
